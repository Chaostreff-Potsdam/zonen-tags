```python
from PIL import Image, ImageDraw, ImageFont
```

```python
RED = (255, 0, 0, 255)
WHITE = (255, 255, 255, 255)
BLACK = (0, 0, 0, 0)

COLORS = [BLACK, RED, WHITE]
```

```python
img_path = "tag_configurator/static/icons/house.png"
img = Image.open(img_path).convert("RGBA")
```

```python
def find_invalid_pixels(img, colors):
    invalid_pixels = []
    for index, pixel in enumerate(img.getdata()):
        if pixel not in colors:
            invalid_pixels.append((index%img.width, index//img.width, pixel))

    invalid_pixels
    return invalid_pixels

invalid_pixels = find_invalid_pixels(img, COLORS)

if len(invalid_pixels):
    print(f"{len(invalid_pixels)} invalid pixels found.")
    print(invalid_pixels, )
    print(find_closest_color(invalid_pixels[0][2], COLORS))
```

```python
def get_pixel_difference(pixel1, pixel2):
    return sum([abs(pixel1[i] - pixel2[i]) for i in range(4)])

def find_closest_color(pixel, colors):
    min_diff = 1000
    closest_color = None
    for color in colors:
        diff = get_pixel_difference(pixel, color)
        if diff < min_diff:
            min_diff = diff
            closest_color = color
    return closest_color

def fix_image(img, colors):
    for index, pixel in enumerate(img.getdata()):
        if pixel not in colors:
            img.putpixel((index%img.width, index//img.width), find_closest_color(pixel, colors))

    return img
```

```python
fixed = fix_image(img, COLORS)
```

```python
fixed.save(img_path)
```
